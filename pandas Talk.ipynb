#### 2. How much money to I spend each day of the week?

```python
# First, identify the days of the week:

weekday_dict = {0: 'Mon', 1: 'Tues', 2: 'Wed', 3: 'Thur', 4: 'Fri', 5: 'Sat', 6: 'Sun'}

def create_wk_day(field):
    """ Convert our field to Timestamp, then apply weekday method
    Monday == 0 ... Sunday == 6"""
    dt = pd.to_datetime(field)
    wk_day = dt.weekday()
    return weekday_dict[wk_day]

data['weekday'] = data['date'].apply(create_wk_day)

print data.head()
```

```python
# Let's also look at Tues
print data[(data.weekday == 'Tues') | (data.weekday == 'Wed')].head(10)
```

```python
# Next, Group by days of the week

grouped = data['amount'].groupby(data['weekday'])

print grouped

# Creates a GroupBy object that has all the information to apply some operation to each of the groups
```

```python
print "Average Transaction is:"
print grouped.mean()
print 
print "Total spend by day is:"
print grouped.sum()
```

```python
# Iterate over the groups
for name, group in grouped:
    print name
    print group.sum()
    print
```

## Other Stuff to Know:

```python
# 1. Merge (default is inner join)
df1 = pd.DataFrame({'id': [1,2,3,4], 'col2':[11,12,13,14]} )
df2 = pd.DataFrame({'id': [1,2,3,4], 'col3': ['lions', 'tigers', 'bears', 'o my']})

df3 = pd.merge(df1, df2)
print df3
```

```python
# 2. Left join (you can specify left, right, inner, outer
df4 = pd.DataFrame({'new_id':[1,2,3,5,6,7,8], 'col4':[9,9,9,9,9,9,9]})

df5 = pd.merge(df3, df4, left_on='id', right_on='new_id', how='left')
print df5
```

```python
#3. Nulls
#Check if there are nulls
print df5['col4'].isnull()
print

#Fill the NaN's with Whatever you want
print df5.fillna(999)
```

```python
#4. Combining / Stacking DataFrames
print df1
print df2

print pd.concat([df1, df2])

#I've used it when using GroupBy and reading in files by chunk
```
